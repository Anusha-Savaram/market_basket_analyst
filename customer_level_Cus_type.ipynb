```python
import pandas as pd
```

```python
df=pd.read_excel('Capstone Workbook_V3.xlsx',sheet_name='Cleaned_Rawdata')
```

```python
df.head()
```

```python
def change_customer_type(cols):
    name1=cols[0]
    name2=cols[1]
    tmp=df[df['customer_id']==name1]
    if 'Returning' in tmp['Customer_level_type'].unique():
        return 'Returning'
    else:
        return 'First-time'
    del tmp
```

```python
df['Customer_level_type2']=df[['customer_id','Customer_level_type']].apply(change_customer_type,axis=1)
```

```python
df.to_excel('Capstone Workbook_V3_1.xlsx',sheet_name='Cleaned_Rawdata1')
```
